Python para análise de dados(Pandas)

```python
#importando a biblioteca pandas
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/gapminder_full.csv") #Colocar entre aspas o caminho do google drive
```

```python
#visualizando as 5 primeiras linhas
#caso o panda não leia corretamente , devemos ir lá em cima e após as ultima aspas colocar que desejamos que o sep=";" , especificando qual será o separador dos dados.
df.head()
```

```python
#Precisamos colocar a própria variável antes para que ela possa ser salva atualizada
df = df.rename(columns={"country":"pais","year":"ano", "population":"populacao", "life_exp": "expectativa de vida", "gdp_cap" : "PIB", "continent": "continente" }) #Remomeando colunas, neste caso estou usando o parâmetro para mudar de Country para Pais.
```

```python
df.head(10) # Neste caso estou informando que desejo ver as primeiras 10 linhas
```

```python
#Como verificar quantas linhas e colunas temos no Dataframe
df.shape
```

```python
#Retornar apenas o nome das colunas
df.columns
```

```python
#Ver o tipo de dados que está armazenado em cada coluna
df.dtypes
```

```python
#Como retornar as últimas linhas, Neste caso estou pedindo as 15 ultimas linhas
df.tail(15)
```

```python
df.describe() #Obter as informações estatisticas do conjunto de dados
```

```python
#Como fazer um filtro para conjunto de dados, Retornar valor apenas de Determinado Ano por exemplo
df["continente"].unique()
```

```python
#Como atribuir os valores do filtro e retornar resultados de apenas uma coluna
oceania = df.loc[df["continente"] == "Oceania"]
oceania.head()
```

```python
oceania["continente"].unique()
```

```python
#Como agrupar por algum filtro, neste exemplo vamos agrupar por continente para saber quantos países temos em cada continente
df.groupby("continente")["pais"].nunique()
```

```python
#Qual a expectativa de vida média para cada pais em determinado ano
df.groupby("ano")["expectativa de vida"].mean()
```

```python
#Qual a média do PIB?
df["PIB"].mean()
```

```python
#Qual a soma do PIB?
df["PIB"].sum()
```

```python
#Criando meu filtro para retornar os dados do brasil
brazil = df.loc[df["pais"] == "Brazil"]
brazil.head()
```

```python
#Retornei o PIB médio do Brasil
brazil["PIB"].mean()
```

```python
#Verificando a expectativa de vida média do Brasil
brazil.groupby("ano")["expectativa de vida"].mean()
```
